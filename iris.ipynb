# Univariate Analysis

```python
df.variety.value_counts(normalize=True)
df.variety.value_counts(normalize=True).plot.barh()
plt.show()
```

```python
df.variety.value_counts(normalize=True)
df.variety.value_counts(normalize=True).plot.pie()
plt.show()
```

# Bivariate Analysis

```python
df.plot.scatter(x='sepal length',y='sepal width')
plt.show()
```

```python
df[['sepal length','sepal width','petal length','petal width']].corr()
sns.heatmap(df[['sepal length','sepal width','petal length','petal width']].corr(),annot=True,cmap='Blues')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="variety",height=7).map(plt.scatter,'sepal length','sepal width').add_legend()
plt.show()
```

```python
sns.pairplot(data=df, vars=['sepal length','sepal width','petal length','petal width'])
plt.show()
```

```python
sns.pairplot(df,hue="variety",height=3)
plt.show()
```
